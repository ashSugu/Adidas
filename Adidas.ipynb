There's something that i've noticed in this data, 
Usually products of different colours with the same product name will recieve different reviews from the audience. Let's look deeper into the data and find out the basis on which the review count segmentation is being done.

Let's look at the unique rating count per product name.

```python
grouped_count_df = df1.groupby('name')['average_rating'].nunique().reset_index(name = 'unique_rating_count')
print(grouped_count_df)
```

```python
grouped_count_df[grouped_count_df['unique_rating_count'] > 1]
```

From the above table, we can clearly see that, for the same product taken into consideration, the customer reviews and ratings are segmented purely based on it's selling price. For a product at a given price, all the available colours, category fall into the same avarage_customer_rating. There can be downsides because of this kind of segmentation , here the customer sentiment on colour and size is not at all considered. But that can have a huge impact on making product specific customer analysis.

```python
plt.plot(df1['average_rating'], df1['selling_price'])
```

Let's plot a bar graph on the product category vs the no of purchases made

This is not a time series data on purchase history, the data we have in hand is just the product list present instock. But how do we atleast come up with a rough estimate on how many would have made a purchase? To tackle this issue, let's make us of the review count section in the data. Let's assume that on an average 1 in 4 customers who made a purchase will write a review. 

```python
df2 = df1.groupby('category')['reviews_count'].sum().reset_index(name = 'total_purchases')
```

```python
df2.head()
```

```python
plt.bar(df2['category'], df2['total_purchases'])
```

```python
plt.pie(df2['total_purchases'], labels = ['Accessories', 'Clothing', 'Shoes'], autopct='%1.1f%%')
plt.show()
```

The data above shows that 92.6% of the Adidas' purchase is made on Shoes!

Let's calculate how much of revenue is made by each of the categories.

```python
df1.head(2)
```

```python
df1['revenue'] = df1['selling_price'] * df1['reviews_count'] * 4
df1.head(2)
```

```python
df3 = df1.groupby('category')['revenue'].sum().reset_index(name = 'total_revenue')
df3.head()
```

```python
plt.bar(df3['category'], df3['total_revenue'])
plt.show()
```

```python
plt.pie(df3['total_revenue'], labels = ['Accessories', 'Clothing', 'Shoes'], autopct='%1.1f%%')
plt.show()
```

We see that 96.1% of Adidas' total revenue is from their shoes!

(Note: We have made an assume that on an average 1 in 4 customers who made a purchase will write a review.)

```python
df4 = df1[df1['category'] == 'Shoes']
```

```python
df4.head(3)
```

```python
df5 = df4.groupby('name')['revenue'].sum().reset_index(name = 'Revenue_per_product')
df5.head()
```

```python
plt.bar(df5['name'], df5['Revenue_per_product'])
```

```python
plt.pie(df5['Revenue_per_product'], autopct='%1.1f%%')
plt.show()
```

```python
max_revenue_row = df5.loc[df5['Revenue_per_product'].idxmax()]
print(max_revenue_row)
```

Among the shoes, the Stan Smith Shoes produce the highest revenue. It accounts for 14.5% of the total revenue generated out of shoes.

## Conclusion

In this analysis, I explored the Adidas US product dataset from Kaggle to uncover review patterns across different products. A key insight emerged: distinct product variants (such as different colors of the same shoe) shared identical review counts and ratings. While initially this seemed like a dataset quirk, further investigation on Adidasâ€™ own website revealed the same pattern reviews were aggregated across all color variants of a product.

This has important implications. Aggregating reviews at the product level may simplify backend systems, but it also risks masking real customer sentiment. For example, if one color variant consistently underperforms in quality or appeal, that signal may be lost entirely. This could mislead both customers and internal teams, affecting decisions in design, marketing, and inventory management.

This analysis suggests that brands like Adidas could benefit from tracking and analyzing reviews at the variant level. A small structural change in their data pipeline could unlock more precise insights, enabling better alignment with customer preferences and improved product strategy.
